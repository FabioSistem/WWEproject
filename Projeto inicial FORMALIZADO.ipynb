## Primeira Análise
Esta análise foi possível obter o número de dias que cada lutador ficou o cinturão e qual cinturão foi pesquisado, nesse caso o WWE CHAMPIONSHIP, baseado nos dados encontrados podemos entender qual lutador esteve mais dias com o cinturão, onde ocorreu as
mudanças, o quão a empresa mudou no passar dos anos, a quantidade de vezes que um lutador conquistou o cinturão, a cidade na qual
o título mudou de mãos desde que foi criado.

```python
reinadowwechampion = Lutadores.loc[Lutadores['belt']== 'WWE Championship']
reinadowwechampion = reinadowwechampion.groupby(['name']).agg({'name':'count', 'days': 'sum'})
reinadowwechampion = reinadowwechampion.sort_values(['days'], ascending=[False])
display(reinadowwechampion)
```

Nesta tabela podemos ver os lutadores que permaneceram mais tempo com o título mundial juntando todas as vezes que venceram e defenderam o mesmo, um destaque maior nessa lista é do lutador JOHN CENA que possui o recorde de reinados da geração atual de 2005 até hoje tendo vencido 13 vezes e contabilizando quase 1300 dias como campeão.

```python
eventwintitles2 = Lutadores.copy().loc[Lutadores['belt'] == 'WWE Championship']
eventwintitles2['event_WrestleMania'] = eventwintitles2['event'].str.contains('WrestleMania')
eventwintitles2['event_arrumado'] = np.where(eventwintitles2['event_WrestleMania'], 'WrestleMania', eventwintitles2['event'])

eventwintitles2['event_SmackDown'] = eventwintitles2['event'].str.contains('SmackDown')
eventwintitles2['event_arrumado'] = np.where(eventwintitles2['event_SmackDown'], 'SmackDown', eventwintitles2['event_arrumado'])

eventwintitles2 = eventwintitles2.groupby(['event_arrumado']).agg({'name': 'count'})
eventwintitles2 = eventwintitles2.sort_values('name', ascending=False )
display(eventwintitles2)
```

Nesta tabela analisamos de forma ordenada e simplificada o evento da empresa que tivemos novos campeões e quantas vezes ocorreu essa mudança. podemos ver no topo da tabela que entre os 4 Pay Per views mais importantes da história destacamos a WRESTLEMANIA com 24 mudanças de título e entre esses eventos o principal show da empresa o MONDAY NIGHT RAW com 14 vezes em que tivemos novos campeões.

```python
locationwintitles = Lutadores.loc[Lutadores['belt']== 'WWE Championship']
locationwintitles = locationwintitles.groupby(['location']).agg({'name': 'count'})
locationwintitles = locationwintitles.sort_values('name', ascending=False )
display(locationwintitles)
```

Por fim, nessa tabela podemos ver a cidade na qual tivemos por mais vezes novos campeões sendo coroados, com destaque para duas das cidades mais conhecidas pelos turistas que visitam anualmente o país norte americano, são elas: Nova Iorque localizada no Estado de New York e Los Angeles localizada no estado da Califórnia.
